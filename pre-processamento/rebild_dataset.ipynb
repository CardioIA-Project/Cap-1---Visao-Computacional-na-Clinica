```python
from pathlib import Path
from sklearn.model_selection import train_test_split
import shutil
```

```python
BASE_DIR = Path(
    r"C:\Users\vitor\Documents\GitHub\Cap-1---Visao-Computacional-na-Clinica\dataset\archive\chest_xray\chest_xray"
)

OUTPUT_DIR = Path(
    r"C:\Users\vitor\Documents\GitHub\Cap-1---Visao-Computacional-na-Clinica\dataset\dataset_rebalanced"
)

CLASSES = ["NORMAL", "PNEUMONIA"]

RANDOM_STATE = 42

# ==========================
# CRIAR ESTRUTURA
# ==========================

for split in ["train", "val", "test"]:
    for classe in CLASSES:
        (OUTPUT_DIR / split / classe).mkdir(parents=True, exist_ok=True)
```

```python
for classe in CLASSES:

    imagens = []

    # junta train + val + test
    for origem_split in ["train", "val", "test"]:

        pasta = BASE_DIR / origem_split / classe

        imagens.extend([
            str(arq)
            for arq in pasta.glob("*")
            if arq.suffix.lower() in [".jpg", ".jpeg", ".png"]
        ])

    print(f"\n{classe}: {len(imagens)} imagens")

    # 70% treino
    train_imgs, temp_imgs = train_test_split(
        imagens,
        test_size=0.30,
        random_state=RANDOM_STATE,
        shuffle=True
    )

    # 15% validação
    # 15% teste
    val_imgs, test_imgs = train_test_split(
        temp_imgs,
        test_size=0.50,
        random_state=RANDOM_STATE,
        shuffle=True
    )

    conjuntos = {
        "train": train_imgs,
        "val": val_imgs,
        "test": test_imgs
    }

    for split, lista in conjuntos.items():

        for arquivo in lista:

            origem = Path(arquivo)

            destino = (
                OUTPUT_DIR
                / split
                / classe
                / origem.name
            )

            shutil.copy2(origem, destino)
```

```python
print("\nDistribuição Final")

for split in ["train", "val", "test"]:

    print(f"\n{split.upper()}")

    for classe in CLASSES:

        total = len(
            list(
                (OUTPUT_DIR / split / classe).glob("*")
            )
        )

        print(f"{classe}: {total}")
```
